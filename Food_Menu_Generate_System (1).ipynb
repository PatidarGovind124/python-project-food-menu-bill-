```python
# Function to add a new item to the menu
def add_item(menu):
    name = input("Enter the name of the new item: ")
    Price = float(input("Enter the price of the new item: "))
    Description = input("Enter the description of the new item: ")
    new_item = {"name": name, "Price": Price, "Description": Description}
    menu.append(new_item)
    print(f"{name} has been added to the menu")
```

```python
# Function to edit an existing item in the menu
def edit_item(menu):
    view_menu(menu)
    item_idx = int(input("Enter the number of the item you want to edit: ")) - 1
    if 0 <= item_idx < len(menu):
        item = menu[item_idx]
        print("Current details:")
        print(f"Name: {item['name']}, Price: ₹{item['Price']:.2f}, Description: {item['Description']}")
        item['name'] = input("Enter the new name (leave blank to keep the current name): ") or item['name']
        item['Price'] = float(input("Enter the new price (leave blank to keep the current price): ") or item['Price'])
        item['Description'] = input("Enter the new description (leave blank to keep the current description): ") or item['Description']
        print("Item details updated successfully.")
    else:
        print("Invalid item number. Please try again.")
```

```python

# Function to delete an item from the menu
def delete_item(menu):
    view_menu(menu)
    item_idx = int(input("Enter the number of the item you want to remove: ")) - 1
    if 0 <= item_idx < len(menu):
        deleted_item = menu.pop(item_idx)
        print(f"{deleted_item['name']} has been deleted from the menu.")
    else:
        print("Invalid item number. Please try again.")

def main():
    while True:
        print("\nOptions:")
        print("1. View Menu")
        print("2. Add Item")
        print("3. Edit Item")
        print("4. Delete Item")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            view_menu(menu_items + new_menu_items)
        elif choice == '2':
            add_item(new_menu_items)
        elif choice == '3':
            edit_item(menu_items + new_menu_items)
        elif choice == '4':
            delete_item(menu_items + new_menu_items)
        elif choice == '5':
            print("Exiting the Food menu system. Goodbye and Have A Nice Day!")
            break
        else:
            print("Invalid Choice. Please try again.")

# Call the main function
main()
```
